```python
import pandas as pd

def calculate_hei(data, ideal_values):
    # Initialize an empty list to store HEI values for each row
    hei_values = []

    for _, row in data.iterrows():
        # Calculate the sub-index for each parameter
        sub_indices = [(row[col] / ideal_values[col]) * 100 for col in row.index]

        # Calculate the unit weights for each parameter
        unit_weights = [1 / ideal_values[col] for col in row.index]

        # Calculate the HEI for the current row
        hei_row = sum(unit_weight * sub_index for unit_weight, sub_index in zip(unit_weights, sub_indices))

        # Append the HEI value for the current row to the list
        hei_values.append(hei_row)

    return hei_values

# Load the data
data = pd.read_csv('D:\jupyter\heavy-metals\heavymetal.csv')

# Ideal values for heavy metals in drinking water (mg/L)
ideal_values = {
    'Cadmium': 0.003,
    'Lead': 0.01,
    'Copper': 2.0,
    'Zinc': 3.0,
    'Nickel': 0.2,
    'Chromium': 0.05,
    'Iron': 0.3,
    'Manganese': 0.05 
}

# Calculate the HEI for the first 200 rows
hei_values_first_200_rows = calculate_hei(data.head(383), ideal_values)

print("HEI for the first 200 rows:", hei_values_first_200_rows)
```

```python
import pandas as pd
import numpy as np

def calculate_HQUI(sample):
  HQUI = 0
  for parameter, value in zip(ideal_values.keys(), sample):
    if value > ideal_values[parameter]:
      HQUI += (value - ideal_values[parameter]) ** 2
  return HQUI

def classify_sample(HQUI):
  if HQUI <= 0.1:
    return '1'
  elif HQUI <= 0.2:
    return '1'
  elif HQUI <= 0.3:
    return '1'
  elif HQUI <= 0.4:
    return '0'
  elif HQUI <= 0.5:
    return '0'
  else:
    return '0'

# Ideal values for heavy metals in drinking water (mg/L)
ideal_values = {
  'Cadmium': 0.003,
  'Lead': 0.01,
  'Copper': 2.0,
  'Zinc': 3.0,
  'Nickel': 0.2,
  'Chromium': 0.05,
  'Iron': 0.3,
  'Manganese': 0.1
}

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('D:\jupyter\heavy-metals\heavymetal.csv')

# Calculate the HQUI for each sample
HQUI = []
for sample in data.values:
  HQUI.append(calculate_HQUI(sample))

# Add the HQUI and classification to the DataFrame
data['HQUI'] = HQUI
data['Classification'] = data['HQUI'].apply(classify_sample)

# Print the DataFrame
print(data.to_string())
```
